```python
# FIRST
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt((f"{os.path.abspath('../data')}/populations.txt"))
```

```python
year = data[:, 0]
hares = data[:, 1]
lynxes = data[:, 2]
carrots =  data[:, 3]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(year, hares)
ax.plot(year, lynxes)
ax.plot(year, carrots)

```
