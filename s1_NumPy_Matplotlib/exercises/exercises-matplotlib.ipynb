## Simple Plotting

1. Plot a the sine function from 0 to 6 (inclusinve) using 50 points.
2. Plot the line again in red and with diamond shaped markers of size 5.
3. Now plot the line, but include a legend.

*Helpful functions*: ``plt.plot``, ``plt.legend``, ``plt.grid``

```python
fun = np.sin(np.linspace(0, 6, 50))
fig = plt.figure()
ax = plt.subplot(111)
ax.plot(fun,  color='red', marker='D', markersize=5)
plt.title("Something", color='orange')
ax.legend()
plt.show()
# fig  = plt.plot(fun, color='red', marker='D', markersize=5)
# fig.add_axes(legend)


```
